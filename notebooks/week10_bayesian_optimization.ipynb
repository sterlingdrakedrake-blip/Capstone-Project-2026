## 4. Core GP and Acquisition Functions

GP and UCB functions unchanged. Seed updated to 987.

**CHANGED:** F1 now uses a fixed pre-planned sweep point rather than dynamic max-distance calculation.

````python
def fit_gp(X, y):
    gp = GaussianProcessRegressor(
        kernel=Matern(nu=2.5), n_restarts_optimizer=10, normalize_y=True
    )
    gp.fit(X, y)
    return gp

def suggest_next_query(X, y, n_dims, n_candidates=100000, kappa=2.0, seed=987):
    """CHANGED: seed updated to 987."""
    rng = np.random.default_rng(seed)
    gp = fit_gp(X, y)
    candidates = rng.uniform(0, 1, size=(n_candidates, n_dims))
    mu, sigma = gp.predict(candidates, return_std=True)
    ucb = mu + kappa * sigma
    best_idx = np.argmax(ucb)
    best_x = candidates[best_idx]
    mu_b, sig_b = gp.predict(best_x.reshape(1, -1), return_std=True)
    return best_x, float(mu_b.ravel()[0]), float(sig_b.ravel()[0])

def format_query(x):
    return '-'.join([f'{v:.6f}' for v in np.clip(x, 0.0, 0.999999)])

# CHANGED: F1 now uses pre-planned fixed sweep coordinates instead of dynamic max-distance
# Rationale: 9 rounds of zero signal -- systematic coverage is more principled than
# recalculating the maximum-distance point each round.
F1_SWEEP_PLAN = {
    10: np.array([0.227592, 0.561706]),  # central-left region
    11: np.array([0.272742, 0.308863]),  # lower-left quadrant
    12: np.array([0.227592, 0.950000]),  # upper-left region
    13: np.array([0.591806, 0.239632]),  # lower-center region
}

print('Functions defined.')
print(f'F1 sweep plan: {len(F1_SWEEP_PLAN)} pre-planned coordinates for W10-W13')
````

## 5. Week 10 Adaptive Kappa Strategy

**CHANGED from Week 9:** Major kappa adjustments based on W9 results.

| kappa | Functions | Rationale |
|---|---|---|
| 0.5 | F2, F8 | New all-time bests in W9 -- exploit hard |
| 1.0 | F4 | Positive output in W9, exploit near R4 basin |
| 2.0 | F5 | Declining 3 rounds, moderate re-exploration |
| 2.5 | F3, F6, F7 | Regressed W9, explore new regions |
| Planned | F1 | Fixed sweep coordinate W10 |

````python
# CHANGED: F2 from 2.0->0.5 (new best); F4 from 2.5->1.0 (positive again); F5 from 1.5->2.0; F3 from 1.5->2.5
kappa_per_function = {
    1: None,   # CHANGED: fixed sweep, GP not used
    2: 0.5,    # CHANGED from 2.0 -- new all-time best 0.647 W9
    3: 2.5,    # CHANGED from 1.5 -- regressed W9
    4: 1.0,    # CHANGED from 2.5 -- positive 0.050 W9
    5: 2.0,    # CHANGED from 1.5 -- declining 3 rounds
    6: 2.5,    # unchanged -- regressed W9
    7: 2.5,    # unchanged -- badly regressed W9
    8: 0.5,    # unchanged -- new best 9.979 W9
}

strategy_notes = {
    1: f"Planned sweep W10: {format_query(F1_SWEEP_PLAN[10])} (central-left region)",
    2: "GP-UCB kappa=0.5 (new all-time best 0.647 W9, exploit hard)",
    3: "GP-UCB kappa=2.5 (regressed W9, explore new region)",
    4: "GP-UCB kappa=1.0 (positive 0.050 W9, exploit near R4 basin)",
    5: "GP-UCB kappa=2.0 (declining 3 rounds, moderate re-exploration)",
    6: "GP-UCB kappa=2.5 (regressed W9, explore new region)",
    7: "GP-UCB kappa=2.5 (badly regressed W9 to 0.114, explore new region)",
    8: "GP-UCB kappa=0.5 (new best 9.979 W9, exploit very hard)",
}
print('Kappa assignments set.')
````

## 6. Run Week 10 Queries

````python
CURRENT_ROUND = 10
results = {}

for func_id, d in functions.items():
    X, y = d['X'], d['y']
    n_dims = X.shape[1]
    kappa = kappa_per_function[func_id]

    # CHANGED: F1 uses pre-planned sweep coordinate
    if func_id == 1:
        best_x = F1_SWEEP_PLAN[CURRENT_ROUND]
        gp = fit_gp(X, y)
        mu_b = float(gp.predict(best_x.reshape(1, -1)).ravel()[0])
        sig_b = float(gp.predict(best_x.reshape(1, -1), return_std=True)[1].ravel()[0])
    else:
        best_x, mu_b, sig_b = suggest_next_query(X, y, n_dims, kappa=kappa)

    query_str = format_query(best_x)
    results[func_id] = {
        'query_string': query_str, 'best_x': best_x,
        'current_best_y': float(y.max()), 'predicted_mu': mu_b,
        'predicted_sigma': sig_b, 'kappa': kappa, 'strategy': strategy_notes[func_id],
    }

    print(f"F{func_id} ({n_dims}D) | n={len(y)} | Best Y: {y.max():.6f}")
    print(f"  Strategy : {strategy_notes[func_id]}")
    print(f"  GP mu={mu_b:.4f} | sigma={sig_b:.4f}")
    print(f"  Query    : {query_str}")
    print()
````
